# Creating and Saving an Audio Template

```python
# Import
from cross_correlation import *
```

```python
# Audio path for the species you want to make a template for
path = './data/20190628_115000.WAV'
```

```python
# Load in the audio
audio = load_audio(path)
```

```python
# Visualizes audio
spectrogram(audio)
```

```python
len(audio)
```

```python
# Create start and end for the audio range that contains the template
start = 80000
end = 110000
```

```python
#Creates the template
template = np.abs(librosa.stft(audio[start:end]))
fig, ax = plt.subplots()
img = librosa.display.specshow(librosa.amplitude_to_db(template,ref=np.max),x_axis='time',ax=ax)
plt.title("Template")
```

```python
# Path to save template to, ONLY CHANGE THE SPECIES NAME
template_path = './data/templates/Flycatcher'
```

```python
# Saves template
np.save(template_path, template)
```

```python
# Checks that the template saved correctly
(np.load(template_path + '.npy') == template).mean()
```
